## Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

#Features
X = df.drop('Survived',axis=1)
y = df['Survived']

#Train/test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.3, random_state=42)
```

```python
#initalize and train the model
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)
```

```python
#Predictions
y_pred = lr.predict(X_test)

print("Accuracy:", accuracy_score(y_test,y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test,y_pred))
print("Classification Report:\n", classification_report(y_test,y_pred))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop('Survived',axis=1)
y = df['Survived']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train,y_train)
```

```python
#Predict
y_pred_rf = rf.predict(X_test)

print("Accuracy:", accuracy_score(y_pred_rf,y_test))
print("Confusion Matrix:\n", confusion_matrix(y_pred_rf,y_test))
print("Classification Report:\n", classification_report(y_pred_rf,y_test))
```

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss',random_state=42)
xgb_model.fit(X_train,y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)

print("Accuracy:", accuracy_score(y_pred_xgb, y_test))
print("Classification Matrix:\n", confusion_matrix(y_pred_xgb,y_test))
print("Clasification Report:\n", classification_report(y_pred_xgb,y_test))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svm = SVC()

param_grid = {} # A base version (no tuning)

grid_search = GridSearchCV(svm, param_grid, cv=5, scoring='accuracy',n_jobs=-1)

grid_search.fit(X_train,y_train)

print("Best Parameters:", grid_search.best_params_)
```

```python
y_pred_svm = grid_search.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_svm))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

## Hyperparameter Tuning

```python
#For Logistic Regression first
log_hp = LogisticRegression(max_iter=500)

#Hyperparameters
param_grid_log_hp = {
    'C':[0.01,0.1,1,10,100],
    'penalty':['l1','l2'],
    'solver':['liblinear']
}

grid_search = GridSearchCV(estimator=log_hp,param_grid=param_grid_log_hp,cv=5,scoring='accuracy')

grid_search.fit(X_train,y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```

```python
#Predictions
best_log_hp = grid_search.best_estimator_
y_pred_hp = best_log_hp.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test,y_pred_hp))
print("Confusion Matrix:\n", confusion_matrix(y_test,y_pred_hp))
print(classification_report(y_test,y_pred_hp))
```

## Hyperparameter Tuning for Random Forest

```python
rf_hp_grid = {
    'n_estimators':[100,200], #Number of trees
    'max_depth':[None,5,10], #depth of trees
    'min_samples_split':[2,5], #min samples to split a node
    'min_samples_leaf':[1,2], #min samples at a leaf node
    'bootstrap':[True,False] #whether bootstrap samples are used
}

rf_hp = RandomForestClassifier(random_state=42)
grid_search_hp = GridSearchCV(estimator=rf_hp,param_grid=rf_hp_grid,cv=5,n_jobs=-1,scoring='accuracy')
```

```python
grid_search_hp.fit(X_train,y_train)
```

```python
best_rf = grid_search_hp.best_estimator_
y_pred_rf = grid_search_hp.predict(X_test)

print("Accuracy:", accuracy_score(y_test,y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```

## Cross Validation

```python
#For Logistic Regression

from sklearn.model_selection import cross_val_score
model = LogisticRegression(max_iter=1000)

#5-fold cross-validation
cv_scores = cross_val_score(model,X,y,cv=5,scoring='accuracy')

print("Cross-Validation Scores:", cv_scores)
print("Mean Accuracy:", cv_scores.mean())
```

```python
#For Random Forest
rf_model = RandomForestClassifier(random_state=42)

cv_rf = cross_val_score(rf_model,X,y,cv=5,scoring='accuracy')
print("Random Forest Scores:", cv_rf.mean())
```

```python
#SVM
svm_model = SVC()
cv_svm = cross_val_score(svm_model,X,y,cv=5,scoring='accuracy')
print("SVM Scoring:", cv_svm.mean())
```
